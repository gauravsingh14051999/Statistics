```python
import matplotlib.pyplot as plt
df=pd.read_csv('alphabet_stock_data.csv')
df1=df[['Date','Close']]
df2=np.where((df1['Date']>='2020-04-01') & (df1['Date']<='2020-09-30'))
df3=df1.iloc[df2]
plt.suptitle('Stock prices of Alphabet Inc.,\n01-04-2020 to 30-09-2020', \
                 fontsize=18, color='black')
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("$ price", fontsize=16, color='black')
df3['Close'].plot(color='green')

```

```python
#14.
# Write a Pandas program to create a plot of adjusted closing prices, thirty days and forty days 
# simple moving average of Alphabet Inc.  between two specific dates. 

import matplotlib.pyplot as plt
df=pd.read_csv('alphabet_stock_data.csv')
df1=df[['Date','Adj Close']]
df2=np.where((df1['Date']>='2020-04-01') & (df1['Date']<='2020-09-30'))
df3=df1.iloc[df2]
df3['SMA_30_days'] = df3.iloc[:,1].rolling(window=30).mean() 
df3['SMA_40_days'] = df3.iloc[:,1].rolling(window=40).mean()
plt.figure(figsize=[8,6])
plt.grid(True)
plt.title('Historical stock prices of Alphabet Inc. [01-04-2020 to 30-09-2020]\n',fontsize=18, color='black')
plt.plot(df3['Adj Close'],label='Adjusted Closing Price', color='black')
plt.plot(df3['SMA_30_days'],label='30 days simple moving average', color='red')
plt.plot(df3['SMA_40_days'],label='40 days simple moving average', color='green')
plt.legend(loc=2)
plt.show()
```

```python
#15.Write a Pandas program to create a plot of adjusted closing prices, 30 days simple 
#moving average and exponential moving average of Alphabet Inc. between two specific dates.

import matplotlib.pyplot as plt
df=pd.read_csv('alphabet_stock_data.csv')
df1=df[['Date','Adj Close']]
df2=np.where((df1['Date']>='2020-04-01') & (df1['Date']<='2020-09-30'))
df3=df1.iloc[df2]
df3['30_days']=df3.iloc[:,1].rolling(window=30).mean()
df3['20_days'] = df3.iloc[:,1].ewm(span=20,adjust=False).mean()

plt.grid(True)
plt.title('Historical stock prices of Alphabet Inc. [01-04-2020 to 30-09-2020]\n',fontsize=18, color='black')
plt.plot(df3['Adj Close'],label='Adjusted Closing Price', color='blue')
plt.plot(df3['30_days'],label='30 days simple moving average', color='red')
plt.plot(df3['20_days'],label='20 days exponential moving average', color='green')
plt.legend(loc=2)
plt.show()
```

```python
#16. Write a Pandas program to create a scatter plot of the trading volume/stock prices of Alphabet Inc. 
#stock between two specific dates.

import matplotlib.pyplot as plt
df=pd.read_csv('alphabet_stock_data.csv')
df1=df[['Close','Volume']]
df2=np.where((df['Date']>='2020-04-01') & (df['Date']<='2020-09-30'))
df3=df1.iloc[df2]
x=['Close']
y=['Volume']
df3.plot.scatter(x,y);

plt.grid(True)
plt.title('Trading Volume/Price of Alphabet Inc. stock,\n01-04-2020 to 30-09-2020', fontsize=14, color='black')
plt.xlabel("Stock Price",fontsize=12, color='black')
plt.ylabel("Trading Volume", fontsize=12, color='black') 
plt.show()
```
